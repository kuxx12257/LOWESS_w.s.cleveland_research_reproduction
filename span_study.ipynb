```python
#AIM: TO observe the differences in smootheness of the fit on changing the values of the span (f)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#OUR GENERATED DATSET
np.random.seed(0)

x=np.linspace(0,10,100) #equally spaced data points
y=np.e**(np.sin(x)+np.cos(x))+np.random.normal(0,0.3,100) #the dependent variable values with noise introduced in it
```

```python
#FITTING THE LOWESS model
from statsmodels.nonparametric.smoothers_lowess import lowess
fractions = [0.1,0.2,0.4, 0.5,0.8]#here we take up few sample values of the span to observe its effect on the model prediction

plt.scatter(x, y)

for f in fractions:
    smoothed = lowess(y, x, frac=f)
    plt.plot(smoothed[:,0], smoothed[:,1], label=f"f={f}")

plt.legend()
plt.title("Effect of Span (f)")
plt.show()

#from the plot as we can see the value of span vriable is quite crucial
#a significantly large value of the span may lead to underfitting of the model and might not fit the datset in the desired manner
#a smaller value of span variable may lead to overfitting of the outliers as well aling withh the relevant data and hece is not a good approximation
#hence the span value must be perfect or optimal not to laarge andnot too small to prevent both underfitting and overfitting of the model
```
