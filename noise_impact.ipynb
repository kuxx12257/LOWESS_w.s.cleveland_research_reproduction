```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
%matplotlib inline
```

```python
#OUR GENERATED DATSET
np.random.seed(0)

x=np.linspace(0,10,100) #equally spaced data points
y=np.e**(np.sin(x)+np.cos(x))+np.random.normal(0,0.3,100) #the dependent variable values with noise introduced in it
```

```python
for noise in [0.1,0.3,0.5,1.0]:
  y_noise=np.e**(np.sin(x)+np.cos(x))+np.random.normal(0,noise,100)
  reg=sm.nonparametric.lowess(y_noise,x,frac=0.3)

  plt.scatter(x,y_noise)
  plt.plot(reg[:,0],reg[:,1],color="red")
  plt.title(f"effect of noise on the dataset {noise}")
  plt.grid(True)
  plt.show()
  #from the plot we cannsee that on incresing the amount of noise in the dataset the smoothness of the predicted plot slowly decreases
  #but a point to note is that our model works quite well with lower noise and almost fits prefectly and smoothly
  #but when noise becomes too much slowly the smoothness of the fit reduces
```
