```python
#AIM: NON LINEAR NOISY DATSET CREATION AND TO SHOW HOW LINEAR REGRESSION FAILS
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
#NOISY AND NONLINEAR DATASET CREATION
np.random.seed(0)

x=np.linspace(0,10,100) #equally spaced data points
y=np.e**(np.sin(x)+np.cos(x))+np.random.normal(0,0.3,100) #the dependent variable values with noise introduced in it
```

```python
#plot of the generated dataset
plt.scatter(x,y)
plt.title("noisy data")
plt.show()
```

```python
#trying to fit the datset with a linear regression model
from sklearn import linear_model
regr=linear_model.LinearRegression()
regr.fit(x.reshape(-1,1),y)
y_predicted=regr.predict(x.reshape(-1,1))
```

```python
#plotting the preddiction made by linear regression
plt.scatter(x,y)
plt.plot(x,y_predicted,color="green")
plt.title("prediction according to linear regression")
plt.show()
#from the plot its quite evident that linear regression fails badly in case of non linear noisy data
```

```python
from sklearn.metrics import r2_score
print("r2=",r2_score(y,y_predicted.reshape(-1,1)))
#from r2_score also we can easily see that linear regression fails to fit the datset
```
