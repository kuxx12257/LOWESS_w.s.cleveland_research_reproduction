```python
#NOW WE WILL TRY TO FIT THE LOWESS model to ouur artificiallly generated datset:
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#OUR GENERATED DATSET
np.random.seed(0)

x=np.linspace(0,10,100) #equally spaced data points
y=np.e**(np.sin(x)+np.cos(x))+np.random.normal(0,0.3,100) #the dependent variable values with noise introduced in it
```

```python
#FITTING THE LOWESS model
from statsmodels.nonparametric.smoothers_lowess import lowess

lowess_result=lowess(y,x,frac=0.3)#initially we are setting the frac value to 0.3
x_lowess=lowess_result[:,0]
y_lowess=lowess_result[:,1]
```

```python
plt.scatter(x,y)
plt.plot(x_lowess,y_lowess,color="red")
plt.title("fitting the lowess model")
plt.show()
#from the plot we can clearly see that th lowess model fits well with our datset
```

```python
from sklearn.metrics import r2_score
print("r2_score=",r2_score(y_lowess,y))
#r2_score also reflects that the lowess model is better fit to our dataset
```
